```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Iris.csv")
data.sample(5)
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.pairplot(data, hue='Species', height=2.5)
plt.suptitle('All Relationships in Iris Dataset', y=1.02)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Sepal Length vs Sepal Width
axes[0, 0].scatter(data['SepalLengthCm'], data['SepalWidthCm'], c='blue', alpha=0.6)
axes[0, 0].set_xlabel('Sepal Length (cm)')
axes[0, 0].set_ylabel('Sepal Width (cm)')
axes[0, 0].set_title('Sepal Length vs Width')

# Petal Length vs Petal Width
axes[0, 1].scatter(data['PetalLengthCm'], data['PetalWidthCm'], c='red', alpha=0.6)
axes[0, 1].set_xlabel('Petal Length (cm)')
axes[0, 1].set_ylabel('Petal Width (cm)')
axes[0, 1].set_title('Petal Length vs Width')

# Sepal Length vs Petal Length
axes[1, 0].scatter(data['SepalLengthCm'], data['PetalLengthCm'], c='green', alpha=0.6)
axes[1, 0].set_xlabel('Sepal Length (cm)')
axes[1, 0].set_ylabel('Petal Length (cm)')
axes[1, 0].set_title('Sepal Length vs Petal Length')

# Sepal Width vs Petal Width
axes[1, 1].scatter(data['SepalWidthCm'], data['PetalWidthCm'], c='purple', alpha=0.6)
axes[1, 1].set_xlabel('Sepal Width (cm)')
axes[1, 1].set_ylabel('Petal Width (cm)')
axes[1, 1].set_title('Sepal Width vs Petal Width')

plt.tight_layout()
plt.show()
```

1. Sepal length and weight: **No Correlation**
2. petal length and weight: **Strong Correlation**
3. sepal length and petal length: **Strong Correlation**
4. sepal width and petal width: **No Correlation**

```python
def histogram_generator(col_name: str, data=data):
    plt.figure(figsize=(10, 6))
    sns.histplot(data[col_name], bins=20)
    plt.xlabel(col_name)
    plt.ylabel('Frequency (Number of Flowers)')
    plt.title(f'Distribution of {col_name} in Iris Flowers')
    plt.show()
```

```python
histogram_generator("SepalLengthCm")
```

* Presence of distinct iris species, as shown by the histogram data distribution (3 peaks were identified 5.0, 5.7, 6.4)

```python
histogram_generator("SepalWidthCm")
```

* appears as a slightly right skew, indicating the sepal is wider in distinct subgroup.

```python
histogram_generator("PetalLengthCm")
```

* appears that the petal lenght is distinctly bimodal with one group, indicating the petal lenght is diffrent in diffrent species.

```python
histogram_generator("PetalWidthCm")
```

* appears that the petal width is distinctly multimodal, indicating the petal width is the best feature for distinguishing between different iris species.
